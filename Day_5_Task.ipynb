## Average Temperature Calculation

```python
df["avg_temp"] = (df["temp_max"] + df["temp_min"]) / 2
```

```python
overall_avg_temp = df["avg_temp"].mean()
overall_avg_temp
```

## Detect Extreme Weather Days

```python
extreme_hot_days = df[df["temp_max"] > 35]
extreme_cold_days = df[df["temp_min"] < 5]
```

```python
print("Extreme Hot Days:", extreme_hot_days.shape[0])
print("Extreme Cold Days:", extreme_cold_days.shape[0])
```
