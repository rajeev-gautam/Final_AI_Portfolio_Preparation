# **Numerical Computing Toolkit with NumPy**

## Importing Important Libraries

```python
import numpy as np
import time
```

## Vector Operations Using Loops

```python
def vec_add(a, b):
    res = []
    for i in range(len(a)):
        res.append(a[i] + b[i])
    return res

def vec_sub(a, b):
    res = []
    for i in range(len(a)):
        res.append(a[i] - b[i])
    return res

def vec_elem_mul(a, b):
    res = []
    for i in range(len(a)):
        res.append(a[i] * b[i])
    return res

def vec_dot(a, b):
    total = 0
    for i in range(len(a)):
        total += a[i] * b[i]
    return total

def vec_normalize(a):
    total = 0
    for x in a:
        total += x * x
    mag = total ** 0.5

    res = []
    for x in a:
        res.append(x / mag)
    return res
```

```python
v1 = [1, 2, 3, 4, 5]
v2 = [6, 7, 8, 9, 10]

vec_add(v1, v2), vec_sub(v1, v2), vec_elem_mul(v1, v2), vec_dot(v1, v2), vec_normalize(v1)
```

## Vector Operations using Numpy

```python
def vec_add_np(a, b): return a + b
def vec_sub_np(a, b): return a - b
def vec_elem_mul_np(a, b): return a * b
def vec_dot_np(a, b): return np.dot(a, b)
def vec_normalize_np(a): return a / np.linalg.norm(a)
```

```python
v1_np = np.array(v1)
v2_np = np.array(v2)

vec_add_np(v1_np, v2_np), vec_sub_np(v1_np, v2_np), vec_elem_mul_np(v1_np, v2_np), vec_dot_np(v1_np, v2_np), vec_normalize_np(v1_np)
```

## Matrix Operations using Loops

```python
def mat_add(A, B):
    res = []
    for i in range(len(A)):
        row = []
        for j in range(len(A[0])):
            row.append(A[i][j] + B[i][j])
        res.append(row)
    return res

def mat_sub(A, B):
    res = []
    for i in range(len(A)):
        row = []
        for j in range(len(A[0])):
            row.append(A[i][j] - B[i][j])
        res.append(row)
    return res

def mat_multiply(A, B):
    res = []
    for i in range(len(A)):
        row = []
        for j in range(len(B[0])):
            total = 0
            for k in range(len(B)):
                total += A[i][k] * B[k][j]
            row.append(total)
        res.append(row)
    return res

def mat_transpose(A):
    res = []
    for j in range(len(A[0])):
        row = []
        for i in range(len(A)):
            row.append(A[i][j])
        res.append(row)
    return res
```

```python
A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]

mat_add(A, B), mat_sub(A, B), mat_multiply(A, B), mat_transpose(A)
```

## Matrix Operations using Numpy

```python
def mat_add_np(A, B): return A + B
def mat_sub_np(A, B): return A - B
def mat_multiply_np(A, B): return A @ B
def mat_transpose_np(A): return A.T
```

```python
A_np = np.array(A)
B_np = np.array(B)

mat_add_np(A_np, B_np), mat_sub_np(A_np, B_np), mat_multiply_np(A_np, B_np), mat_transpose_np(A_np)
```

## Comparing Python Loops VS Numpy Performance (Vector Operations)

```python
size = 300_000
a = list(range(size))
b = list(range(size))

a_np = np.array(a)
b_np = np.array(b)

def time_it(func, *args):
    start = time.time()
    func(*args)
    return time.time() - start

print("VECTOR PERFORMANCE\n")

print("Add:", time_it(vec_add, a, b), "|", time_it(vec_add_np, a_np, b_np))
print("Sub:", time_it(vec_sub, a, b), "|", time_it(vec_sub_np, a_np, b_np))
print("Elem Mul:", time_it(vec_elem_mul, a, b), "|", time_it(vec_elem_mul_np, a_np, b_np))
print("Dot:", time_it(vec_dot, a, b), "|", time_it(vec_dot_np, a_np, b_np))
print("Normalize:", time_it(vec_normalize, a), "|", time_it(vec_normalize_np, a_np))
```

## Comparing Python Loops VS Numpy Performance (Matrix Operations)

```python
n = 400
A = [[i for i in range(n)] for _ in range(n)]
B = [[i for i in range(n)] for _ in range(n)]

A_np = np.array(A)
B_np = np.array(B)

print("\nMATRIX PERFORMANCE\n")

print("Add:", time_it(mat_add, A, B), "|", time_it(mat_add_np, A_np, B_np))
print("Sub:", time_it(mat_sub, A, B), "|", time_it(mat_sub_np, A_np, B_np))
print("Multiply:", time_it(mat_multiply, A, B), "|", time_it(mat_multiply_np, A_np, B_np))
print("Transpose:", time_it(mat_transpose, A), "|", time_it(mat_transpose_np, A_np))
```
